```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Carregar a imagem
img = cv2.imread('../../assets/pentagono.webp', cv2.IMREAD_COLOR)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# Aplicar a deteco de bordas Canny
edges = cv2.Canny(gray, 50, 150, apertureSize=3)

# Aplicar a Transformada de Hough para detectar linhas
lines = cv2.HoughLines(edges, 1, np.pi / 180, 200)

# Desenhar as linhas detectadas na imagem original
if lines is not None:
    for rho, theta in lines[:, 0]:
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a * rho
        y0 = b * rho
        x1 = int(x0 + 1000 * (-b))
        y1 = int(y0 + 1000 * (a))
        x2 = int(x0 - 1000 * (-b))
        y2 = int(y0 - 1000 * (a))
        cv2.line(img, (x1, y1), (x2, y2), (0, 0, 255), 2)

# Mostrar os resultados
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.imshow(cv2.cvtColor(edges, cv2.COLOR_BGR2RGB))
plt.title('Deteco de Bordas')
plt.axis('off')

plt.subplot(122)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Transformada de Hough')
plt.axis('off')

plt.show()
```
